# Question 1: Bias and Variance (**10 pts**)


(a). (**4 pts**) Describe the difference between model bias and the bias of a point estimator.
> While a model uses input data to predict a value as its output, a point estimator is a statistic representing an unknown population parameter. Note that a point estimator can be mean, variance, and even a regression coefficient. Now, model bias refers to the inaccuracy in the predictive model's ouput. It represents how far the model's prediction was from the true value of a given input. Meanwhile, the bias of a point estimator is the error between the point estimator itself (the predicted parameter value) and the true value of the parameter. It measures whether the estimator is an over or underestimate of the parameter of interest. Hence, while the model bias is a reflection of the accuracy of entire predictive model and how well it approximates the underlying data generation, the bias of a point estimator corresponds to a one specific statistical measure and describes how closely the point estimator approximates the true parameter value.

(b) (**6 pts**). How can you use a learning curve to determine whether a model is overfitting (for a given sample size)? Discuss this with respect to the observed train and validation error curves. How does your answer change if the model you are trying to determine is underfitting?
> The typical learning curve shows a change in some error metric (e.g., RMSE for regression) as the training set size changes. We expect the training set error to increase and then converge to the true error as we increase the training set size. Meanwhile, the test set error decrease and converge to the true error as the training set size increases. Now, when there is a large gap between the train and test error at a given sample size, this most likely indicates our model is overfitting. The training error will be especially low compared to the test error, which will be rather high, exacerbating the gap between the two curves. Meanwhile, if both the training and test error are high, this indicates the model is underfitting, as it is unable to learn the patterns of the underlying data.

Consider a function $g$ such that $g(x)=2sin(\pi

a) (2 points) Generate a dataset $(D = X,Y)$ by generating 2000 data points in the following way -
*  $x_i = \frac{i}{2000}$ for each $i$ in $1, 2, 3 \ldots 2000$.
* Then, for each of the $x_i$'s obtain $y_i = 2sin(\pi x_i)cos(3\pi x_i^2)+ N(0,1)$ where $N(0,1)$ denotes the normal distribution with mean 0 and variance 1.

**Generate 10 datasets** by independently repeating the entire process 10 times. Plot any one of the generated datasets.

```python
import numpy as np
import math
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
def generate_data(num_points):
  x = np.arange(1, num_points + 1) / num_points
  ### START CODE ###
  # Compute y according to the labeling function g
  mean = 0
  stddev = 1 # variance is sqrt(stddev) = 1
  N = np.random.normal(mean, stddev, num_points)
  y = [2 * math.sin(math.pi * i) * math.cos(3 * math.pi * (i ** 2)) + N[j] for j, i in enumerate(x)]
  ### END CODE ###
  return x,y
```

```python
datasets = {}
for j in range(10):
  datasets[j] = {}
  datasets[j]['X'], datasets[j]['Y'] = generate_data(2000)

### START CODE ###
# Plot y v.s. x of any one of the dataset
plt.scatter(datasets[0]['X'], datasets[0]['Y'], s=2);
### END CODE ###
```

b) (5 points) Consider an estimator of function $g$, $f(x)$ that divides the x's into $K$ bins such that there are $m = n/K$ data points in each bin and the predicted y for all points in that bin is the mean of all the $y$'s in the bin.
$$
f(x) = \sum_{j=1}^{K} \bar{y}_j \mathbb{1}(x \in \text{bin } j)
$$
where
$$
\bar{y}_j = \frac{1}{m}\sum_{i=j*m}^{(j+1)*m-1} y_i.
$$
Implement this estimator. Obtain $Y_{hat} = f(x)$ for one of the datasets created above by using $K=25$ and plot the predicted $Y_{hat}$ along with $X$ and $Y$.

```python
### START CODE ###
import statistics as stats
def estimator(x, y, K):
    m = int(len(x)/K)
    xbins = [x[(j*m):(j+1)*m] for j in range(K)]
    ybins = [y[(j*m):(j+1)*m] for j in range(K)]
    fx = []
    
    for j, xbin in enumerate(xbins):
        for x in xbin:
            fx.append(stats.mean(ybins[j]))

    return fx
### END CODE ###
```

```python
### START CODE ###
# Plot predicted y_{hat} along with X and Y
yhat = estimator(datasets[0]['X'], datasets[0]['Y'], 25)
plt.scatter(datasets[0]['X'], datasets[0]['Y'], label='$Y$', s=2);
plt.plot(datasets[0]['X'], yhat, label='$Y_{hat}$', color='tomato');
plt.legend();
### END CODE ###
```

c) (3 points) What do you think will happen to the MSE, the bias and the variance as the number of bins $K$ is increased?

> **MSE** will decrease since, as $K$ approaches the number of data points in $X$, the estimator will begin to return predictions that exactly match the values of samples in the data. Since the estimator is effectively a windowed average, increasing the number of bins (i.e., decreasing the bin or window size $m$) results in the "estimation" converging toward the exact value of each point in the dataset: a bin size of $m=1$ (when $K = 2000$) is the same as returning $Y$. Similarly, **bias** will decrease because the smaller window size will "smooth out" the data less, resulting in a noisier estimation. For the same reason, **variance** will increase since the estimation will begin to more closely resemble the noisiness in the underlying data. In each case, this is akin to overfitting the estimator on the observations. This behavior is visualized below by plotting $Y_{hat}$ with $K=2000$. The estimation at this maximum value of $K$ is indiscernable from the underlying observations.

```python
yhat = estimator(datasets[0]['X'], datasets[0]['Y'], 2000)
plt.scatter(datasets[0]['X'], datasets[0]['Y'], label='$Y$', s=2);
plt.plot(datasets[0]['X'], yhat, label='$Y_{hat}$', color='tomato');
plt.legend();
```

```python
mse_scores = []
K = [2,5,10,25,50,100,250,500,1000]
yhats = {}
for n in K:
  ### START CODE ###
  # compute y_hat given n for each dataset
  yhats[n] = [estimator(datasets[i]['X'], datasets[i]['Y'], n) for i in range(len(datasets))]
  ### END CODE ###

  ### START CODE ###
  # Compute MSE for chosen dataset
  mse = mean_squared_error(datasets[0]['Y'], yhats[n][0])
  ### END CODE ###
  mse_scores.append(mse)
plt.plot(K,mse_scores)
plt.show()
```

We are using the term "sample" to remember that bias/variance are "expected quantities" that we are approximating by considering only 10 models, one per dataset.

```python
bias_vals = []
var_vals = []
K = [2,5,10,25,50,100,250,500,1000]
pred_y = np.zeros((len(datasets[0]['Y']),10)) # axis 0: number of data points, axis 1: number of datasets (10)

for n in K:
  for j in datasets.keys():
    x = datasets[j]['X']
    y = datasets[j]['Y']
    ### START CODE ###
    # Estimate y_{hat}
    y_hat = estimator(x, y, n)
    ### END CODE ###
    pred_y[:,j] = y_hat

  ### START CODE ###
  # Compute the mean of the estimates over 10 datasets for each input x_{i}
  avg_yhat = np.average(pred_y, axis=1) # 1D array of shape (2000,)
  # Compute the MSE between avg_yhat and y as bias
  true_y = [2 * math.sin(math.pi * i) * math.cos(3 * math.pi * (i ** 2)) for i in datasets[0]['X']]
  bias = mean_squared_error(avg_yhat, true_y)
  # Compute the variance of the estimates over 10 datasets for each input x_{i}, then average over all inputs
  var_yhat = np.average(np.var(pred_y, axis=1))
  ### END CODE ###
  bias_vals.append(bias)
  var_vals.append(var_yhat)
plt.plot(K, bias_vals, label='bias')
plt.plot(K, var_vals, label='variance')
plt.legend()
plt.show()
```

## Part 1. (**10 pts**) Stochastic gradient descent derivation ##

Use stochastic gradient descent to derive the coefficent updates (assuming squared loss is being used as the cost function) for the 4 coefficients $w_0, w_1, w_2, w_3$ in this model：

$ y = w_0 + w_1e^{-x_1} + w_2 x_1 + w_3x_1x_2 $

> A coefficient update in SGD can be formalized as follows:

$
w' = w - \eta \frac{\partial \ell}{\partial w}
$

> Where $w$ is a coefficient, $\eta$ is the learning rate, and $\ell$ is the loss function. Our first step is to compute the partial derivative of the loss function $\ell$ with respect to each coefficient $w$. The squared loss function for inputs $x$ and target $t$ is defined as follows:

$
\ell = (t - y_w(x))^2
$

> Given this loss function, we can compute the partial derivatives:

$\frac{\partial\ell}{\partial w_0} = \frac{\partial}{\partial w_0}(t - y_w(x))^2 = -2(t - y_w(x))\frac{\partial y_w}{\partial w_0}$

$\frac{\partial\ell}{\partial w_1} = -2(t - y_w(x))\frac{\partial y_w}{\partial w_1}$

$\frac{\partial\ell}{\partial w_2} = -2(t - y_w(x))\frac{\partial y_w}{\partial w_2}$

$\frac{\partial\ell}{\partial w_3} = -2(t - y_w(x))\frac{\partial y_w}{\partial w_3}$

> Next, we need to take the partial derivative of the model with respect to each coefficient:

$
\frac{\partial y}{\partial w_0} = \frac{\partial}{\partial w_0}w_0 + \frac{\partial}{\partial w_0}w_1 e^{-x_1} + \frac{\partial}{\partial w_0}w_2 x_1 + \frac{\partial}{\partial w_0}w_3 x_1 x_2 = 1 + 0 + 0 + 0 = 1
$

$
\frac{\partial y}{\partial w_1} = \frac{\partial y}{\partial w_1}w_0 + \frac{\partial y}{\partial w_1}w_1e^{-x_1} + \frac{\partial y}{\partial w_1}w_2 x_1 + \frac{\partial y}{\partial w_1}w_3 x_1 x_2 = 0 + (1)(e^{-x_1}) + 0 + 0 = e^{-x_1}
$

$
\frac{\partial y}{\partial w_2} = \frac{\partial y}{\partial w_2}w_0 + \frac{\partial y}{\partial w_2}w_1e^{-x_1} + \frac{\partial y}{\partial w_2}w_2 x_1 + \frac{\partial y}{\partial w_2}w_3x_1x_2 = 0 + 0 + (1)(x_1) + 0 = x_1
$

$
\frac{\partial y}{\partial w_3} = \frac{\partial y}{\partial w_3}w_0 + \frac{\partial y}{\partial w_3}w_1e^{-x_1} + \frac{\partial y}{\partial w_3}w_2 x_1 + \frac{\partial y}{\partial w_3}w_3x_1x_2 = 0 + 0 + 0 + (1) (x_1 x_2) = x_1 x_2
$


> Substituting these into the SGD function, we can compute the update for each coefficient:

$ w_0' = w_0 + 2\eta(t - y_w(x))\frac{\partial y_w}{\partial w_0} = w_0 + 2\eta(t - y_w(x))(1) = w_0 + 2\eta(t - y_w(x))$

$ w_1' = w_1 + 2\eta(t - y_w(x))e^{-x_1}$

$ w_2' = w_2 + 2\eta(t - y_w(x))x_1$

$ w_3' = w_3 + 2\eta(t - y_w(x))x_1x_2$

In this exercise, you are going to visualize four batch update steps of gradient descent for a  linear regression model with two parameters (i.e. weights, indicated by $\theta$).

The true target function is $t = \theta_{0} + \theta_{1}x$ with $\theta_{0}=2$ and $\theta_{1}=0.5$.

Try the following two initializations:

and try the following three learning rates:

Therefore, there will be **six** combinations or settings to consider in total. For each setting, you will plot (a) the data and the changing linear regression fit and (b) the model parameters moving in the weight space after every update.

```python
# Generate data
np.random.seed(42)
m = 20
theta0_true = 2
theta1_true = 0.5
x = np.linspace(-1,1,m)
y = theta0_true + theta1_true * x + np.random.normal(0, 0.2, size=x.shape)
```

```python
def loss_func(theta0, theta1):
    theta0 = np.atleast_3d(np.asarray(theta0))
    theta1 = np.atleast_3d(np.asarray(theta1))
    return np.average((y - model(x, theta0, theta1))**2, axis=2)/2

def model(x, theta0, theta1):
    return theta0 + theta1 * x
```

```python
# Experiment with different initialization and learning rate combinations
### START CODE ###
init_list = [[0, 0], [0, -4]]
lr_list = [0.5, 1, 2.1]
### END CODE ###

# Left column shows the data and the changing linear regression models
# Right column shows the model parameters moving over the loss landscape
fig, ax = plt.subplots(nrows=len(init_list) * len(lr_list), ncols=2, figsize=(12, 36))

for i, init in enumerate(init_list):
    for j, lr in enumerate(lr_list):
        row_idx = i * len(lr_list) + j
        ax[row_idx][0].scatter(x, y, marker='x', s=40, color='k')
        theta0_grid = np.linspace(-1,5,101)
        theta1_grid = np.linspace(-5,5,101)
        L_grid = loss_func(theta0_grid[np.newaxis,:,np.newaxis],
                           theta1_grid[:,np.newaxis,np.newaxis])

        # A labeled contour plot for the right column
        X, Y = np.meshgrid(theta0_grid, theta1_grid)
        contours = ax[row_idx][1].contour(X, Y, L_grid, 30)
        ax[row_idx][1].clabel(contours)
        # The target parameter values indicated on the loss function contour plot
        ax[row_idx][1].scatter([theta0_true]*2,[theta1_true]*2,s=[50,10], color=['k','w'])

        # Take N = 4 steps with learning rate alpha down the steepest gradient, starting at init
        N = 4
        theta = [init] # placeholder list for storing historical parameters
        L = [loss_func(*theta[0])[0]] # placeholder list for storing historical loss values
        for _ in range(N):
            last_theta = theta[-1]
            this_theta = np.empty((2,))
            # Update theta
            ### START CODE ### (2 lines of code)
            this_theta[0] = last_theta[0] - lr * (-1/len(x) * np.sum(2 * (y - (last_theta[0] + last_theta[1] * x))))
            this_theta[1] = last_theta[1] - lr * (-1/len(x) * np.sum(2 * x * (y - (last_theta[0] + last_theta[1] * x))))
            ### END CODE ###
            theta.append(this_theta)
            L.append(loss_func(*this_theta))

        # Annotate the loss function plot with coloured points indicating the
        # parameters chosen and red arrows indicating the steps down the gradient.
        # Also plot the fit function on the LHS data plot in a matching colour.
        colors = ['b', 'g', 'm', 'c', 'orange']
        ax[row_idx][0].plot(x, model(x, *theta[0]), color=colors[0], lw=2,
                   label=r'$\theta_0 = {:.3f}, \theta_1 = {:.3f}$'.format(*theta[0]))
        for k in range(1,N+1):
            ax[row_idx][1].annotate('', xy=theta[k], xytext=theta[k-1],
                           arrowprops={'arrowstyle': '->', 'color': 'r', 'lw': 1},
                           va='center', ha='center')
            ax[row_idx][0].plot(x, model(x, *theta[k]), color=colors[k], lw=2,
                   label=r'$\theta_0 = {:.3f}, \theta_1 = {:.3f}$'.format(*theta[k]))
        ax[row_idx][1].scatter(*zip(*theta), c=colors, s=40, lw=0)

        # Labels and titles.
        ax[row_idx][1].set_xlabel(r'$\theta_0$')
        ax[row_idx][1].set_ylabel(r'$\theta_1$')
        ax[row_idx][1].set_title(f'Loss function (Init:[{init[0]},{init[1]}], LR:{lr})')
        ax[row_idx][0].set_xlabel(r'$x$')
        ax[row_idx][0].set_ylabel(r'$y$')
        ax[row_idx][0].set_title(f'Data and Fit (Init:[{init[0]},{init[1]}], LR:{lr})')
        axbox = ax[row_idx][0].get_position()

plt.show()
```
